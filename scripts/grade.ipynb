```python
import os
import json
import pandas as pd

base_path = "/home/v-yuanteli/mlebench-office/mle-bench/runs_aide_baseline"
output_file = "/home/v-yuanteli/mlebench-office/mle-bench/runs_aide_baseline/res_gpt_4o_aide.csv"

data_list = []

for folder in os.listdir(base_path):
    folder_path = os.path.join(base_path, folder)
    if os.path.isdir(folder_path):
        grading_file_path = None
        for file in os.listdir(folder_path):
            if file.endswith("grading_report.json"):
                grading_file_path = os.path.join(folder_path, file)
                break

        with open(grading_file_path, "r") as f:
            grading_data = json.load(f)

            for report in grading_data.get("competition_reports", []):
                data_list.append({
                    "competition_id": report.get("competition_id"),
                    "score": report.get("score"),
                    "gold_threshold": report.get("gold_threshold"),
                    "silver_threshold": report.get("silver_threshold"),
                    "bronze_threshold": report.get("bronze_threshold"),
                    "median_threshold": report.get("median_threshold"),
                    "any_medal": report.get("any_medal"),
                    "gold_medal": report.get("gold_medal"),
                    "silver_medal": report.get("silver_medal"),
                    "bronze_medal": report.get("bronze_medal"),
                    "above_median": report.get("above_median"),
                    "submission_exists": report.get("submission_exists"),
                    "valid_submission": report.get("valid_submission"),
                })

competition_df = pd.DataFrame(data_list)
competition_df
```

```python
grouped_df = competition_df.groupby('competition_id').agg(
    total_competitions=('competition_id', 'size'),
    valid_submissions=('valid_submission', 'sum'),
    submission_exists=('submission_exists', 'sum'),
    any_medal=('any_medal', 'sum'),
    gold_medal=('gold_medal', 'sum'),
    silver_medal=('silver_medal', 'sum'),
    bronze_medal=('bronze_medal', 'sum'),
    above_median=('above_median', 'sum')
).reset_index()
# grouped_df.to_csv("/home/v-yuanteli/mlebench-office/mle-bench/runs_aide_baseline/res_gpt_4o_aide_each.csv", index=False)
grouped_df
```

```python
import pandas as pd
metrics = ["any_medal", "gold_medal", "silver_medal", "bronze_medal", "above_median", "submission_exists", "valid_submission"]
df = pd.read_csv('aide_gpt_4o_our_results.csv')
df['relative_time'] = pd.to_datetime(df['times']) - pd.to_datetime(df['times'].min())

def compute_ratio_within_time(df, metric, hours):
    df_time = df[df['relative_time'] <= pd.Timedelta(hours=hours)]
    if df_time.empty:
        return 0

    last_entries = df_time.groupby('competition').apply(lambda x: x.loc[x['relative_time'].idxmax()])
    return last_entries[metric].sum() / last_entries.shape[0] if last_entries.shape[0] > 0 else 0  # 避免除零错误

aide_ratios_11h = {metric: round(compute_ratio_within_time(df, metric, 11) * 100, 1) for metric in metrics}
aide_ratios_11h
```

```python
import numpy as np

# 计算总比赛次数
total_competitions = filtered_df['total_competitions'].sum()

# 计算各项比率
metrics = {
    "comp_made_rate": filtered_df['submission_exists'].sum() / total_competitions,
    "comp_finished_rate": filtered_df['valid_submissions'].sum() / total_competitions,
    "above_median_rate": filtered_df['above_median'].sum() / total_competitions,
    "gold_medal_rate": filtered_df['gold_medal'].sum() / total_competitions,
    "silver_medal_rate": filtered_df['silver_medal'].sum() / total_competitions,
    "bronze_medal_rate": filtered_df['bronze_medal'].sum() / total_competitions,
    "any_medal_rate": filtered_df['any_medal'].sum() / total_competitions,
}

# 计算标准误差（SE）: SE = sqrt(p(1-p)/n)
se_metrics = {
    key: np.sqrt(value * (1 - value) / total_competitions) for key, value in metrics.items()
}

# 输出计算结果，格式为 “均值 ± 标准误差”
for key, value in metrics.items():
    se = se_metrics[key]
    print(f"{key}: {value:.6f} ± {se:.6f}")

markdown_table = f"""
| AIDE (OpenAI, gpt-4o_2024-08-06) | {comp_made_rate*100:.1f}             | {comp_finished_rate*100:.1f}               | {above_median_rate*100:.1f}           | {bronze_medal_rate*100:.1f}      | {silver_medal_rate*100:.1f}      | {gold_medal_rate*100:.1f}    | {any_medal_rate*100:.1f}         |
"""

```
